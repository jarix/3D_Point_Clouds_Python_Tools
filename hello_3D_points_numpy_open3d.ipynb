# Hello 3D Points with Numpy and Open3D

Hello world for 3D points in an array

```python
# Prerequisites:
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

print("Python version: ", sys.version)
```

### Generate some Random Points

```python
points_3D = np.random.rand(1000, 3)  # 1000 random points in 3D
```

### Plot 3D points with Numpy Scatter Plot

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points_3D[:,0], points_3D[:,1], points_3D[:,2], c='r', marker='o')
ax.set_xlabel('X')
ax.set_ylabel('Y') 
ax.set_zlabel('Z')
plt.title('3D Scatter Plot of Random Points')
plt.show()
```

### Visualize points with Open3D

```python
import open3d as o3d
```

```python
# Create a point cloud object
point_cloud = o3d.geometry.PointCloud()
point_cloud.points = o3d.utility.Vector3dVector(points_3D)

point_cloud.paint_uniform_color([1, 0, 0])  # red color
```

```python
# Visualize 
o3d.visualization.draw_geometries([point_cloud], window_name="3D Point Cloud", width=800, height=600)
```

```python
# Create a visualizer  
vis = o3d.visualization.Visualizer()

# Create a window and add the point cloud
vis.create_window()
vis.add_geometry(point_cloud)

# Chagge point size
opt = vis.get_render_option()
opt.point_size = 10.0
# Set background color
opt.background_color = np.asarray([0.9, 0.9, 0.9])  # gray color
# Set camera parameters
camera = vis.get_view_control()
camera.set_lookat([0, 0, 0])
camera.set_front([0, 0, -1])
camera.set_up([0, 1, 0])
camera.set_zoom(0.8)
# Set light parameters
light = vis.get_render_option().light_on


vis.run()
vis.destroy_window()
```
